```python
def analyze2images(image1, image2):
    h1, w1 = image1.shape[:2]
    h2, w2 = image2.shape[:2]
    
    new_camera_matrix1, _ = cv.getOptimalNewCameraMatrix(mtx, dist, (w1, h1), 1, (w1, h1))
    undistorted_image1 = cv.undistort(image1, mtx, dist, None, new_camera_matrix1)
    
    new_camera_matrix2, _ = cv.getOptimalNewCameraMatrix(mtx, dist, (w2, h2), 1, (w2, h2))
    undistorted_image2 = cv.undistort(image2, mtx, dist, None, new_camera_matrix2)
    
    # Convert to grayscale
    gray1 = cv.cvtColor(undistorted_image1, cv.COLOR_BGR2GRAY)
    gray2 = cv.cvtColor(undistorted_image2, cv.COLOR_BGR2GRAY)
    
    # Apply edge detection
    edges1 = cv.Canny(gray1, threshold1=50, threshold2=150)
    edges2 = cv.Canny(gray2, threshold1=50, threshold2=150)
    
    # ORB feature detector
    orb = cv.SIFT_create()
    
    # Detect keypoints and descriptors
    keypoints1, descriptors1 = orb.detectAndCompute(edges1, None)
    keypoints2, descriptors2 = orb.detectAndCompute(edges2, None)
    
    # Match features using BFMatcher
    bf = cv.BFMatcher(cv.NORM_L2, crossCheck=True)
    matches = bf.match(descriptors1, descriptors2)
    
    # Sort matches by distance (lower is better)
    matches = sorted(matches, key=lambda x: x.distance)
    
    # Extract matched feature coordinates
    matched_features = []
    for match in matches:
        # Query and train indices
        idx1 = match.queryIdx
        idx2 = match.trainIdx
    
        # Get keypoint coordinates
        x1, y1 = keypoints1[idx1].pt  # Coordinates in image1
        x2, y2 = keypoints2[idx2].pt  # Coordinates in image2
    
        # Append to list
        matched_features.append({
            "Image1_X": x1,
            "Image1_Y": y1,
            "Image2_X": x2,
            "Image2_Y": y2,
            "Distance": match.distance
        })
    
    # Convert to Pandas DataFrame
    df = pd.DataFrame(matched_features)
    
    
    # Optionally visualize matches
    matched_image = cv.drawMatches(image1, keypoints1, image2, keypoints2, matches[:50], None, flags=cv.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
    
    # cv.namedWindow('Matched Features', cv.WINDOW_NORMAL)
    # cv.imshow('Matched Features', matched_image)
    # cv.waitKey(0)
    # cv.destroyAllWindows()
    df['Delta_X'] = df['Image2_X'] - df['Image1_X']
    df['Delta_Y'] = df['Image2_Y'] - df['Image1_Y']
    
    # Calculate the averages
    average_delta_x = df['Delta_X'].mean()
    average_delta_y = df['Delta_Y'].mean()
    
    # Add the magnitude of the distance as an extra metric (optional)
    df['Distance_Magnitude'] = (df['Delta_X']**2 + df['Delta_Y']**2)**0.5
    return df
```

```python
def estimate_feature_distance(df, delta_d, focal_length):
    """
    Estimate the real-world distance to a feature using the shift in pixel coordinates.
    
    Parameters:
        df (pd.DataFrame): DataFrame containing Delta_X and Delta_Y columns.
        delta_d (float): Distance between the two camera positions (meters).
        focal_length (float): Focal length of the camera (in pixels).
    
    Returns:
        pd.Series: Estimated distances to each feature.
    """
    # Compute the total pixel shift for each feature
    df['Delta_p'] = (df['Delta_X']**2 + df['Delta_Y']**2)**0.5
    
    # Estimate the distance to each feature
    df['Estimated_Distance'] = (delta_d * focal_length) / df['Delta_p']
    
    return df[['Delta_X', 'Delta_Y', 'Delta_p', 'Estimated_Distance']]

```

```python

# Function to estimate distances for 20 random points and display them on the image
def visualize_distances(df, image_path, delta_d, focal_length):
    """
    Randomly selects 20 points, calculates distances, and overlays them on the image.
    
    Parameters:
        df (pd.DataFrame): DataFrame containing matched feature coordinates (Image1_X, Image1_Y, Delta_X, Delta_Y).
        image_path (str): Path to the image file.
        delta_d (float): Real-world distance between camera positions (e.g., meters).
        focal_length (float): Focal length of the camera (e.g., in pixels).
    """
    # Select 20 random points
    random_points = df.sample(n=20, replace=True).copy()

    # Compute the total pixel shift
    random_points['Delta_p'] = (random_points['Delta_X']**2 + random_points['Delta_Y']**2)**0.5

    # Estimate the real-world distance for the selected points
    random_points['Estimated_Distance'] = (delta_d * focal_length) / random_points['Delta_p']

    # Load the image
    image = cv.imread(image_path)
    if image is None:
        raise FileNotFoundError(f"Image not found at {image_path}")

    # Overlay the points and distances on the image
    for _, row in random_points.iterrows():
        x, y = int(row['Image1_X']), int(row['Image1_Y'])
        distance = row['Estimated_Distance']

        # Draw the point on the image
        cv.circle(image, (x, y), radius=5, color=(0, 255, 0), thickness=-1)

        # Add the distance text
        cv.putText(image, f"{distance:.2f}m", (x + 10, y - 10),
                    fontFace=cv.FONT_HERSHEY_SIMPLEX, fontScale=2, color=(255, 255, 255), thickness=3)

    # Display the image
    cv.namedWindow('Feature Distances', cv.WINDOW_NORMAL)
    cv.imwrite('DistanceFeatureskyticka.jpg', image)

    cv.imshow('Feature Distances', image)
    cv.waitKey(0)
    cv.destroyAllWindows()


```

```python
# Example parameters
delta_d = 0.5
focal_length = 3460  
image_path = path1
df = analyze2images(image1, image2)
# Run the visualization function
visualize_distances(df, image_path, delta_d, focal_length)
```
